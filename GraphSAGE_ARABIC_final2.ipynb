```python
!pip install networkx
# Step 2: Import Libraries
import torch
import torch.nn.functional as F
from torch_geometric.nn import SAGEConv
from torch_geometric.data import Data
from torch_geometric.utils import from_networkx
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from transformers import AutoTokenizer, AutoModel
from sklearn.model_selection import train_test_split
import pandas as pd
import networkx as nx
from sklearn.metrics.pairwise import cosine_similarity

# Step 3: Upload Dataset
from google.colab import files
uploaded = files.upload()  # Upload your CSV file (e.g., dataset.csv)

# Load the dataset
df = pd.read_excel(pd.ExcelFile(list(uploaded.keys())[0]), header=0) # Expecting two columns: 'tweets' and 'labels' where label is 0 or 1
df = df.sample(frac=0.1, random_state=42)
df = df[['tweet', 'label']]

# Step 4: Preprocess Text Data and Create Embeddings
tokenizer = AutoTokenizer.from_pretrained('xlm-roberta-base')
model = AutoModel.from_pretrained('xlm-roberta-base')




def get_embeddings(texts):
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings

# Generate embeddings for all texts
embeddings = get_embeddings(df['tweet'].tolist())
labels = torch.tensor(df['label'].values, dtype=torch.long)

# Step 5: Create Graph Using NetworkX
G = nx.Graph()
# Add nodes with features
for i in range(len(df)):
    G.add_node(i, x=embeddings[i])

# Step 6: Create Edges Using Cosine Similarity
similarity_matrix = cosine_similarity(embeddings)
threshold = 0.8  # Define a threshold for creating edges

# Add edges based on similarity
for i in range(len(similarity_matrix)):
    for j in range(i + 1, len(similarity_matrix)):
        if similarity_matrix[i, j] > threshold:
            G.add_edge(i, j)

# Convert NetworkX graph to PyTorch Geometric Data
data = from_networkx(G)
data.x = torch.stack([data.x[i] for i in range(len(G.nodes))])
data.y = labels


```

```python
labeled_idx, test_idx = train_test_split(range(len(df)), test_size=0.2, stratify=df['label'])
labeled_idx, unlabeled_idx = train_test_split(labeled_idx, test_size=0.75, stratify=df['label'].iloc[labeled_idx])

data.labeled_mask = torch.tensor(labeled_idx, dtype=torch.long)
data.unlabeled_mask = torch.tensor(unlabeled_idx, dtype=torch.long)
data.test_mask = torch.tensor(test_idx, dtype=torch.long)
```

```python
import torch.nn as nn

# Modify the GraphSAGE model to include BatchNorm and Dropout
class GraphSAGE(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GraphSAGE, self).__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.bn1 = nn.BatchNorm1d(hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, out_channels)
        self.dropout = nn.Dropout(0.5)  # Adding dropout

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.dropout(x)  # Applying dropout
        x = self.conv2(x, edge_index)
        return x

# Updated learning rate and added weight decay
student_model = GraphSAGE(in_channels=embeddings.size(1), hidden_channels=128, out_channels=2)
teacher_model = GraphSAGE(in_channels=embeddings.size(1), hidden_channels=128, out_channels=2)
optimizer = torch.optim.Adam(student_model.parameters(), lr=0.01, weight_decay=1e-4)
optimizer1 = torch.optim.Adam(teacher_model.parameters(), lr=0.01, weight_decay=1e-4)

# Learning rate scheduler to dynamically adjust learning rate
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

# Update EMA Function for smoother parameter updating
def update_teacher_model(teacher, student, alpha=0.995):  # Slightly higher alpha for smoother updates
    for teacher_param, student_param in zip(teacher.parameters(), student.parameters()):
        teacher_param.data.mul_(alpha).add_(1 - alpha, student_param.data)

# Improved training loop
def train_student():
    student_model.train()
    optimizer.zero_grad()
    out = student_model(data.x, data.edge_index)
    loss = F.cross_entropy(out[data.labeled_mask], data.y[data.labeled_mask])
    loss.backward()
    nn.utils.clip_grad_norm_(student_model.parameters(), max_norm=1.0)  # Gradient clipping
    optimizer.step()
    return loss.item()

def train_teacher():
    teacher_model.train()
    optimizer1.zero_grad()
    out = teacher_model(data.x, data.edge_index)
    loss = F.cross_entropy(out[data.labeled_mask], data.y[data.labeled_mask])
    loss.backward()
    nn.utils.clip_grad_norm_(teacher_model.parameters(), max_norm=1.0)  # Gradient clipping
    optimizer1.step()
    return loss.item()

# Combined loss function for consistency and labeled loss
def train_with_combined_loss():
    student_model.eval()
    teacher_model.eval()

    student_out = student_model(data.x, data.edge_index)
    teacher_out = teacher_model(data.x, data.edge_index)

    student_probs = F.softmax(student_out[data.unlabeled_mask], dim=1)
    teacher_probs = F.softmax(teacher_out[data.unlabeled_mask], dim=1)
    consistency_loss = F.mse_loss(student_probs, teacher_probs)

    student_model.train()
    optimizer.zero_grad()
    total_loss = consistency_loss  # Adjusting weight of consistency loss
    total_loss.backward()
    nn.utils.clip_grad_norm_(student_model.parameters(), max_norm=1.0)
    optimizer.step()
    return total_loss.item()

# Train the models with the new setup
for epoch in range(1, 31):  # Reduced to 31 epochs for faster convergenceepoch
    if epoch==1:
        train_teacher()
    labeled_loss = train_student()
    combined_loss = train_with_combined_loss()
    # update_teacher_model(teacher_model, student_model)
    teacher_model=student_model
    scheduler.step()  # Adjust learning rate

    if epoch % 5 == 0:  # More frequent evaluation
        print(f'Epoch {epoch}, Labeled Loss: {labeled_loss:.4f}, Combined Loss: {combined_loss:.4f}')
```

```python
# Updated test function to evaluate model performance on test data

def test():
    teacher_model.eval()  # Set the teacher model to evaluation mode
    out = teacher_model(data.x, data.edge_index)  # Forward pass through the teacher model
    pred = out.argmax(dim=1)  # Get the class with the highest score for each node
    test_pred = pred[data.test_mask]  # Predictions for the test nodes
    test_true = data.y[data.test_mask]  # True labels for the test nodes

    # Calculate evaluation metrics
    acc = accuracy_score(test_true.cpu(), test_pred.cpu())
    f1 = f1_score(test_true.cpu(), test_pred.cpu(), average='weighted')  # Using weighted average for imbalanced data
    precision = precision_score(test_true.cpu(), test_pred.cpu(), average='weighted')
    recall = recall_score(test_true.cpu(), test_pred.cpu(), average='weighted')

    return acc, f1, precision, recall
# Evaluate the model performance
acc, f1, precision, recall = test()
print(f'Accuracy: {acc:.4f}, F1 Score: {f1:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}')
```
